```python
import pandas as pd
import matplotlib.pyplot as plt
population_data = pd.read_csv('population.csv', skiprows=4)
countries = ['United States', 'India', 'China', 'Brazil', 'Nigeria', 'Germany', 'Japan']
year = '2023'
filtered_population = population_data[population_data['Country Name'].isin(countries)]
plt.figure(figsize=(10, 6))
plt.bar(filtered_population['Country Name'], filtered_population[year], color='green')
plt.xlabel('Country')
plt.ylabel(f'Population in {year}')
plt.title(f'Population of Selected Countries in {year}')
plt.xticks(rotation=360)
plt.show()
```
